# CELL 1.  IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# CELL 2 . LOADING THE DATASET AND SAMPLE CREATION

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("MarketingCampaign she code dataset.csv")

#  Display first rows
print("FIRST 5 ROWS")
display(df.head())

# Dataset shape
print("\nDATASET SHAPE:")
print(df.shape)

# Column names
print("\nCOLUMN NAMES:")
print(df.columns)
```

# CELL 3. DATA CLEANING AND INTERGRATION

```python

# 1. Check data types
print("DATA TYPES:")
display(df.dtypes)

# 2. Check missing values
print("\nMISSING VALUES:")
display(df.isnull().sum())

# 3. Check duplicate records
duplicates = df.duplicated().sum()
print("\nNUMBER OF DUPLICATES:", duplicates)

# Remove duplicates if any
df = df.drop_duplicates()

# 4. Standardize column names
df.columns = df.columns.str.strip().str.replace(" ", "_")

print("\nUPDATED COLUMN NAMES:")
print(df.columns)

# 5. Convert date column to datetime
df["Effective_To_Date"] = pd.to_datetime(df["Effective_To_Date"])

print("\nDATE COLUMN CONVERTED SUCCESSFULLY")

# 6. Final dataset info after cleaning
print("\nFINAL DATASET INFO:")
df.info()
```

# CELL 4. EXPLORATORY DATA ANALYSIS

```python
df.describe()
```

```python
# Summary statistics output
summary = {
    "Total Rows": len(df),
    "Response Rate (%)": (df['Response'] == 'Yes').mean() * 100,
    "Missing Values": df.isnull().sum().sum(),
    "Avg CLV": df['Customer_Lifetime_Value'].mean()
}
print(summary)
```

```python

# 3. Define features for the visualization
num_features = ['Customer_Lifetime_Value', 'Income', 'Monthly_Premium_Auto', 'Total_Claim_Amount']
cat_features = ['State', 'EmploymentStatus', 'Education', 'Response']

fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(22, 30))
axes = axes.flatten()

# Distribution of the  NUMERICAL columns.
for i, col in enumerate(num_features):
    sns.histplot(df[col], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f' Distribution of {col}', fontsize=14)

# Distribution of the Categorical columns
for i, col in enumerate(cat_features):
    idx = i + 4
    order = df[col].value_counts().index
    sns.countplot(data=df, x=col, order=order, hue=col, palette='Set2', ax=axes[idx], legend=False)
    axes[idx].set_title(f'Distribution of {col}', fontsize=14)
    if col in ['Education', 'EmploymentStatus']:
        axes[idx].tick_params(axis='x', rotation=30)

# Response by the numerical columns.
for i, col in enumerate(num_features):
    idx = i + 8
    sns.boxplot(data=df, x='Response', y=col, hue='Response', palette='pastel', ax=axes[idx], legend=False)
    axes[idx].set_title(f' {col} by Response', fontsize=14)

# Marketing Performance
sns.countplot(data=df, x='Sales_Channel', hue='Response', palette='viridis', ax=axes[12])
axes[12].set_title('Response by Sales Channel', fontsize=14)

sns.countplot(data=df, x='Renew_Offer_Type', hue='Response', palette='viridis', ax=axes[13])
axes[13].set_title('Response by Renew Offer Type', fontsize=14)

# Correlation Heatmap (1 plot)
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", ax=axes[14], cbar=False)
axes[14].set_title(' Correlation Matrix', fontsize=14)

plt.tight_layout()
plt.savefig('eda_grid_layout.png')
```

# CELL 5. SUMMARY REPORT GENERATION


```python

# General Metrics
total_customers = len(df)
overall_response_rate = (df['Response'] == 'Yes').mean() * 100
avg_clv = df['Customer_Lifetime_Value'].mean()
total_revenue_monthly = df['Monthly_Premium_Auto'].sum()

# Segment Performance
top_state = df['State'].value_counts().idxmax()
best_channel = df[df['Response'] == 'Yes']['Sales_Channel'].value_counts().idxmax()
best_offer = df[df['Response'] == 'Yes']['Renew_Offer_Type'].value_counts().idxmax()

# Claim Correlation
luxury_claim_avg = df[df['Vehicle_Class'].isin(['Luxury Car', 'Luxury SUV'])]['Total_Claim_Amount'].mean()
standard_claim_avg = df[~df['Vehicle_Class'].isin(['Luxury Car', 'Luxury SUV'])]['Total_Claim_Amount'].mean()

# --- PRINTING THE EXECUTIVE SUMMARY ---

print("="*60)
print("              MARKETING CAMPAIGN EXECUTIVE SUMMARY")
print("="*60)

print(f"""
1. KEY PERFORMANCE INDICATORS (KPIs):
   - Total Customer Base: {total_customers:,}
   - Overall Campaign Response Rate: {overall_response_rate:.2f}%
   - Average Customer Lifetime Value (CLV): ${avg_clv:,.2f}
   - Monthly Premium Revenue: ${total_revenue_monthly:,.2f}

2. VISUAL INSIGHTS & FINDINGS:
   - DOMINANT MARKET: {top_state} is the largest market segment.
   - CHANNEL EFFICIENCY: The '{best_channel}' channel is the most successful at converting customers.
   - OFFER SUCCESS: '{best_offer}' is the most accepted renewal offer.
   - RISK ANALYSIS: Luxury vehicles average ${luxury_claim_avg:,.2f} in claims,
     significantly higher than standard vehicles (${standard_claim_avg:,.2f}).
   - INCOME CORRELATION: There is a high concentration of 'Yes' responses in
     the lower-to-middle income brackets ($20k - $50k).

3. STRATEGIC RECOMMENDATIONS:
   - OPTIMIZE CHANNELS: Increase budget allocation for the '{best_channel}' channel,
     as it shows the highest conversion potential.
   - REVISE PREMIUMS: Review premium rates for Luxury Vehicles to offset the
     disproportionately high claim amounts observed in the boxplots.
   - TARGETED OFFERS: Phase out underperforming offers (Offer 3 & 4) and
     expand the rollout of {best_offer}.
   - RETENTION FOCUS: Focus on the 'Employed' segment with Bachelor's degrees,
     as they represent the highest volume of high-CLV customers.
""")
print("="*60)
```
